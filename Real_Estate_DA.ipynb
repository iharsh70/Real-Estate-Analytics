```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('/content/RealEstate_Dataset.csv')

print("=" * 80)
print("DATA CLEANING AND TRANSFORMATION")
print("=" * 80)

# Create a copy for cleaning
df_clean = df.copy()

print("\n1. CONVERTING PRICE TO NUMERIC")
print("-" * 80)
print(f"Original Price Column Sample:\n{df_clean['Price'].head(10)}")
df_clean['Price_Clean'] = df_clean['Price'].str.replace('$', '').str.replace(',', '').astype(float)
print(f"\nPrice after cleaning - Statistics:")
print(f"Min: ${df_clean['Price_Clean'].min():,.2f}")
print(f"Max: ${df_clean['Price_Clean'].max():,.2f}")
print(f"Mean: ${df_clean['Price_Clean'].mean():,.2f}")

print("\n\n2. CONVERTING BEDROOMS TO NUMERIC")
print("-" * 80)
print(f"Original Bedrooms Column Sample:\n{df_clean['Bedrooms'].head(10)}")
df_clean['Bedrooms_Clean'] = df_clean['Bedrooms'].str.extract('(\d+)')[0].astype(float)
print(f"\nBedrooms after cleaning - Unique values: {sorted(df_clean['Bedrooms_Clean'].dropna().unique())}")

print("\n\n3. CONVERTING BATHROOMS TO NUMERIC")
print("-" * 80)
print(f"Original Bathrooms Column Sample:\n{df_clean['Bathrooms'].head(10)}")
df_clean['Bathrooms_Clean'] = df_clean['Bathrooms'].str.extract('(\d+)')[0].astype(float)
print(f"\nBathrooms after cleaning - Unique values: {sorted(df_clean['Bathrooms_Clean'].dropna().unique())}")

print("\n\n4. CONVERTING AREA (SQFT) TO NUMERIC")
print("-" * 80)
print(f"Original Area Column Sample:\n{df_clean['Area (Sqft)'].head(10)}")
df_clean['Area_Sqft_Clean'] = df_clean['Area (Sqft)'].str.replace(' sqft', '').astype(float)
print(f"\nArea after cleaning - Statistics:")
print(f"Min: {df_clean['Area_Sqft_Clean'].min():,.0f} sqft")
print(f"Max: {df_clean['Area_Sqft_Clean'].max():,.0f} sqft")
print(f"Mean: {df_clean['Area_Sqft_Clean'].mean():,.0f} sqft")

print("\n\n5. CONVERTING LOT SIZE (SQFT) TO NUMERIC")
print("-" * 80)
print(f"Original Lot Size Column Sample:\n{df_clean['Lot Size'].head(10)}")
df_clean['Lot_Size_Sqft_Clean'] = df_clean['Lot Size'].str.replace(' sqft', '').astype(float)
print(f"\nLot Size after cleaning - Statistics:")
print(f"Min: {df_clean['Lot_Size_Sqft_Clean'].min():,.0f} sqft")
print(f"Max: {df_clean['Lot_Size_Sqft_Clean'].max():,.0f} sqft")
print(f"Mean: {df_clean['Lot_Size_Sqft_Clean'].mean():,.0f} sqft")

print("\n\n6. STANDARDIZING CATEGORICAL COLUMNS")
print("-" * 80)
print(f"Status values: {df_clean['Status'].unique()}")
print(f"Property Type values: {df_clean['Property Type'].unique()}")
print(f"State values: {df_clean['State'].unique()}")
print(f"Cities: {df_clean['City'].unique()}")

# Convert to lowercase for consistency
df_clean['Status'] = df_clean['Status'].str.strip()
df_clean['Property_Type'] = df_clean['Property Type'].str.strip()
df_clean['City'] = df_clean['City'].str.strip()
df_clean['State'] = df_clean['State'].str.strip()

print("\n\nData Type Summary After Cleaning:")
print(f"Price_Clean: {df_clean['Price_Clean'].dtype}")
print(f"Bedrooms_Clean: {df_clean['Bedrooms_Clean'].dtype}")
print(f"Bathrooms_Clean: {df_clean['Bathrooms_Clean'].dtype}")
print(f"Area_Sqft_Clean: {df_clean['Area_Sqft_Clean'].dtype}")
print(f"Lot_Size_Sqft_Clean: {df_clean['Lot_Size_Sqft_Clean'].dtype}")

print("\n\nFinal Cleaned Dataset Info:")
print(f"Shape: {df_clean.shape}")
print(f"Total Records: {len(df_clean)}")
print(f"Missing Values in Cleaned Numeric Columns: {df_clean[['Price_Clean', 'Bedrooms_Clean', 'Bathrooms_Clean', 'Area_Sqft_Clean']].isnull().sum().sum()}")

# Save cleaned dataframe for next analysis
df_clean.to_csv('/content/RealEstate_Dataset.csv', index=False)
print("\n✓ Cleaned dataset saved for analysis")
```

```python

import pandas as pd
import numpy as np

# Load cleaned data
df = pd.read_csv('/content/RealEstate_Dataset.csv')

print("=" * 80)
print("LOCATION-BASED ANALYSIS")
print("=" * 80)

print("\n1. CITY-LEVEL ANALYSIS")
print("-" * 80)

city_analysis = df.groupby('City').agg({
    'Price_Clean': ['count', 'mean', 'median', 'min', 'max', 'std'],
    'Days on Market': 'mean',
    'Area_Sqft_Clean': 'mean',
    'Bedrooms_Clean': 'mean',
    'Bathrooms_Clean': 'mean'
}).round(2)

city_analysis.columns = ['Listings', 'Avg_Price', 'Median_Price', 'Min_Price', 'Max_Price', 'Std_Price', 'Avg_DOM', 'Avg_Area', 'Avg_Beds', 'Avg_Baths']
city_analysis = city_analysis.sort_values('Avg_Price', ascending=False)

print("\nCity Summary (sorted by Average Price):")
print(city_analysis.to_string())

print("\n\n2. CITY RANKING BY KEY METRICS")
print("-" * 80)

print("\nTop 5 Cities by Average Price:")
for idx, (city, row) in enumerate(city_analysis.sort_values('Avg_Price', ascending=False).head(5).iterrows(), 1):
    print(f"  {idx}. {city:20s}: ${row['Avg_Price']:>12,.2f} ({int(row['Listings']):,} listings)")

print("\nTop 5 Cities by Listing Activity:")
for idx, (city, row) in enumerate(city_analysis.sort_values('Listings', ascending=False).head(5).iterrows(), 1):
    print(f"  {idx}. {city:20s}: {int(row['Listings']):>6,} listings")

print("\nTop 5 Cities by Median Price:")
for idx, (city, row) in enumerate(city_analysis.sort_values('Median_Price', ascending=False).head(5).iterrows(), 1):
    print(f"  {idx}. {city:20s}: ${row['Median_Price']:>12,.2f}")

print("\nFastest Selling Cities (Lowest Avg Days on Market):")
for idx, (city, row) in enumerate(city_analysis.sort_values('Avg_DOM', ascending=True).head(5).iterrows(), 1):
    print(f"  {idx}. {city:20s}: {row['Avg_DOM']:>6.1f} days")

print("\n\n3. STATE-LEVEL ANALYSIS")
print("-" * 80)

state_analysis = df.groupby('State').agg({
    'Price_Clean': ['count', 'mean', 'median'],
    'Days on Market': 'mean',
    'Area_Sqft_Clean': 'mean',
}).round(2)

state_analysis.columns = ['Listings', 'Avg_Price', 'Median_Price', 'Avg_DOM', 'Avg_Area']
state_analysis = state_analysis.sort_values('Avg_Price', ascending=False)

print("\nState Summary (sorted by Average Price):")
print(state_analysis.to_string())

print("\n\n4. PRICE RANGES BY CITY")
print("-" * 80)

for city in city_analysis.index:
    city_data = df[df['City'] == city]['Price_Clean']
    q1 = city_data.quantile(0.25)
    q3 = city_data.quantile(0.75)
    print(f"\n{city}:")
    print(f"  Budget Segment (0-25%):      ${city_data.min():>12,.2f} - ${q1:>12,.2f}")
    print(f"  Mid-Range Segment (25-75%): ${q1:>12,.2f} - ${q3:>12,.2f}")
    print(f"  Luxury Segment (75-100%):   ${q3:>12,.2f} - ${city_data.max():>12,.2f}")

print("\n\n5. PROPERTY TYPE DISTRIBUTION BY CITY")
print("-" * 80)

ptype_by_city = pd.crosstab(df['City'], df['Property_Type'], margins=True)
print(ptype_by_city)

print("\n✓ Location-based analysis complete")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr, spearmanr

# Load cleaned data
df = pd.read_csv('/content/RealEstate_Dataset.csv')

print("=" * 80)
print("PROPERTY FEATURE IMPACT ANALYSIS & CORRELATION")
print("=" * 80)

print("\n1. CORRELATION ANALYSIS")
print("-" * 80)

# Select numeric columns for correlation
numeric_cols = ['Price_Clean', 'Area_Sqft_Clean', 'Bedrooms_Clean', 'Bathrooms_Clean',
                'Lot_Size_Sqft_Clean', 'Year Built', 'Days on Market']

correlation_matrix = df[numeric_cols].corr()

print("\nPearson Correlation Matrix with Price:")
price_corr = correlation_matrix['Price_Clean'].sort_values(ascending=False)
print(price_corr.to_string())

print("\n\nInterpretation of Key Correlations with Price:")
print("-" * 80)
for col, corr_val in price_corr.items():
    if col != 'Price_Clean':
        if abs(corr_val) > 0.7:
            strength = "STRONG"
        elif abs(corr_val) > 0.5:
            strength = "MODERATE"
        elif abs(corr_val) > 0.3:
            strength = "WEAK"
        else:
            strength = "VERY WEAK"

        direction = "POSITIVE" if corr_val > 0 else "NEGATIVE"
        print(f"  {col:25s}: {corr_val:>7.4f} ({strength:12s} {direction})")

print("\n\n2. PRICE VS AREA ANALYSIS")
print("-" * 80)

pearson_corr, pearson_pval = pearsonr(df['Area_Sqft_Clean'], df['Price_Clean'])
spearman_corr, spearman_pval = spearmanr(df['Area_Sqft_Clean'], df['Price_Clean'])

print(f"\nPearson Correlation:  {pearson_corr:.4f} (p-value: {pearson_pval:.2e})")
print(f"Spearman Correlation: {spearman_corr:.4f} (p-value: {spearman_pval:.2e})")
print(f"\nPrice per Square Foot Statistics:")
df['Price_per_Sqft'] = df['Price_Clean'] / df['Area_Sqft_Clean']
print(f"  Mean:     ${df['Price_per_Sqft'].mean():>8.2f}/sqft")
print(f"  Median:   ${df['Price_per_Sqft'].median():>8.2f}/sqft")
print(f"  Std Dev:  ${df['Price_per_Sqft'].std():>8.2f}/sqft")
print(f"  Min:      ${df['Price_per_Sqft'].min():>8.2f}/sqft")
print(f"  Max:      ${df['Price_per_Sqft'].max():>8.2f}/sqft")

print("\n\n3. PRICE VS BEDROOMS ANALYSIS")
print("-" * 80)

bedroom_analysis = df.groupby('Bedrooms_Clean').agg({
    'Price_Clean': ['count', 'mean', 'median', 'std'],
    'Area_Sqft_Clean': 'mean',
    'Bathrooms_Clean': 'mean',
}).round(2)

bedroom_analysis.columns = ['Count', 'Avg_Price', 'Median_Price', 'Std_Price', 'Avg_Area', 'Avg_Baths']
print("\nPrice Statistics by Bedroom Count:")
print(bedroom_analysis.to_string())

print("\n\n4. PRICE VS BATHROOMS ANALYSIS")
print("-" * 80)

bathroom_analysis = df.groupby('Bathrooms_Clean').agg({
    'Price_Clean': ['count', 'mean', 'median', 'std'],
    'Area_Sqft_Clean': 'mean',
    'Bedrooms_Clean': 'mean',
}).round(2)

bathroom_analysis.columns = ['Count', 'Avg_Price', 'Median_Price', 'Std_Price', 'Avg_Area', 'Avg_Beds']
print("\nPrice Statistics by Bathroom Count:")
print(bathroom_analysis.to_string())

print("\n\n5. PRICE BY PROPERTY TYPE")
print("-" * 80)

ptype_analysis = df.groupby('Property_Type').agg({
    'Price_Clean': ['count', 'mean', 'median', 'std', 'min', 'max'],
    'Area_Sqft_Clean': 'mean',
    'Days on Market': 'mean',
    'Bedrooms_Clean': 'mean',
    'Bathrooms_Clean': 'mean',
}).round(2)

ptype_analysis.columns = ['Count', 'Avg_Price', 'Median_Price', 'Std_Price', 'Min_Price', 'Max_Price',
                          'Avg_Area', 'Avg_DOM', 'Avg_Beds', 'Avg_Baths']
ptype_analysis = ptype_analysis.sort_values('Avg_Price', ascending=False)

print("\nProperty Type Analysis (sorted by Average Price):")
print(ptype_analysis.to_string())

print("\n\n6. PRICE BY YEAR BUILT")
print("-" * 80)

# Group by decade
df['Decade'] = (df['Year Built'] // 10 * 10).astype(int)
decade_analysis = df.groupby('Decade').agg({
    'Price_Clean': ['count', 'mean', 'median'],
    'Area_Sqft_Clean': 'mean',
}).round(2)

decade_analysis.columns = ['Count', 'Avg_Price', 'Median_Price', 'Avg_Area']
print("\nPrice Statistics by Decade Built:")
print(decade_analysis.to_string())

print("\n\n7. KEY INSIGHTS ON FEATURE IMPACT")
print("-" * 80)
print(f"\nStrongest Price Correlations:")
print(f"  1. Lot Size:         {correlation_matrix.loc['Lot_Size_Sqft_Clean', 'Price_Clean']:>7.4f}")
print(f"  2. Bedrooms:         {correlation_matrix.loc['Bedrooms_Clean', 'Price_Clean']:>7.4f}")
print(f"  3. Area (Sqft):      {correlation_matrix.loc['Area_Sqft_Clean', 'Price_Clean']:>7.4f}")
print(f"  4. Bathrooms:        {correlation_matrix.loc['Bathrooms_Clean', 'Price_Clean']:>7.4f}")
print(f"  5. Year Built:       {correlation_matrix.loc['Year Built', 'Price_Clean']:>7.4f}")
print(f"  6. Days on Market:   {correlation_matrix.loc['Days on Market', 'Price_Clean']:>7.4f}")

# Save analysis for future use
df.to_csv('/content/RealEstate_Dataset.csv', index=False)
print("\n✓ Property feature impact analysis complete")
```

```python

import pandas as pd
import numpy as np
from scipy import stats

# Load cleaned data
df = pd.read_csv('/content/RealEstate_Dataset.csv')

print("=" * 80)
print("MARKET DEMAND ANALYSIS")
print("=" * 80)

print("\n1. PROPERTY STATUS DISTRIBUTION")
print("-" * 80)

status_counts = df['Status'].value_counts()
status_pct = df['Status'].value_counts(normalize=True) * 100

print("\nMarket Status Breakdown:")
for status in ['Sold', 'Pending', 'For Sale']:
    count = status_counts.get(status, 0)
    pct = status_pct.get(status, 0)
    print(f"  {status:15s}: {count:6,} properties ({pct:5.1f}%)")

print("\n\n2. PROPERTY TYPE DEMAND ANALYSIS")
print("-" * 80)

demand_analysis = df.groupby('Property_Type').agg({
    'Status': lambda x: (x == 'Sold').sum(),
    'Days on Market': ['mean', 'median'],
    'Price_Clean': 'mean',
}).round(2)

demand_analysis.columns = ['Sold_Count', 'Avg_DOM', 'Median_DOM', 'Avg_Price']
demand_analysis['Sold_Pct'] = (demand_analysis['Sold_Count'] / 3000 * 100).round(1)
demand_analysis = demand_analysis.sort_values('Sold_Count', ascending=False)

print("\nDemand by Property Type:")
print(demand_analysis.to_string())

print("\n\n3. DAYS ON MARKET ANALYSIS")
print("-" * 80)

print(f"\nOverall Days on Market Statistics:")
print(f"  Mean:         {df['Days on Market'].mean():>6.1f} days")
print(f"  Median:       {df['Days on Market'].median():>6.1f} days")
print(f"  Mode:         {df['Days on Market'].mode()[0]:>6.0f} days")
print(f"  Std Dev:      {df['Days on Market'].std():>6.1f} days")

print(f"\nDays on Market by Status:")
for status in ['Sold', 'Pending', 'For Sale']:
    status_data = df[df['Status'] == status]['Days on Market']
    print(f"  {status:15s}: {status_data.mean():>6.1f} avg days (median: {status_data.median():>6.0f})")

print(f"\nDays on Market by Property Type:")
dom_by_type = df.groupby('Property_Type')['Days on Market'].agg(['mean', 'median']).sort_values('mean')
for ptype, row in dom_by_type.iterrows():
    print(f"  {ptype:20s}: {row['mean']:>6.1f} avg days (median: {row['median']:>6.0f})")

print("\n\n4. LISTING VELOCITY BY CITY")
print("-" * 80)

city_demand = df.groupby('City').agg({
    'Status': lambda x: (x == 'Sold').sum(),
    'Days on Market': 'mean',
    'Price_Clean': 'count'
}).round(2)

city_demand.columns = ['Sold', 'Avg_DOM', 'Total_Listings']
city_demand['Sell_Through_Rate'] = (city_demand['Sold'] / city_demand['Total_Listings'] * 100).round(1)
city_demand = city_demand.sort_values('Sell_Through_Rate', ascending=False)

print("\nCity Demand Metrics:")
print(city_demand.to_string())

print("\n\n5. MARKET DEMAND INDICATORS")
print("-" * 80)

total_listings = len(df)
sold_listings = (df['Status'] == 'Sold').sum()
pending_listings = (df['Status'] == 'Pending').sum()
forsale_listings = (df['Status'] == 'For Sale').sum()

sell_through_rate = (sold_listings / total_listings) * 100
pending_rate = (pending_listings / total_listings) * 100

print(f"\nMarket Activity Summary:")
print(f"  Total Listings:        {total_listings:,}")
print(f"  Sold:                  {sold_listings:,} ({(sold_listings/total_listings)*100:.1f}%)")
print(f"  Pending:               {pending_listings:,} ({(pending_listings/total_listings)*100:.1f}%)")
print(f"  For Sale:              {forsale_listings:,} ({(forsale_listings/total_listings)*100:.1f}%)")
print(f"  Sell-Through Rate:     {sell_through_rate:.1f}%")
print(f"  Pending Rate:          {pending_rate:.1f}%")

print("\n\n" + "=" * 80)
print("OUTLIER DETECTION ANALYSIS")
print("=" * 80)

print("\n1. PRICE OUTLIERS (Using IQR Method)")
print("-" * 80)

Q1 = df['Price_Clean'].quantile(0.25)
Q3 = df['Price_Clean'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

price_outliers = df[(df['Price_Clean'] < lower_bound) | (df['Price_Clean'] > upper_bound)]

print(f"\nPrice Bounds (IQR Method):")
print(f"  Lower Bound: ${lower_bound:>12,.2f}")
print(f"  Upper Bound: ${upper_bound:>12,.2f}")
print(f"  IQR:         ${IQR:>12,.2f}")

print(f"\nPrice Outliers Detected:")
print(f"  Total Outliers: {len(price_outliers):,} ({(len(price_outliers)/len(df)*100):.1f}%)")
print(f"  Low Outliers:   {len(df[df['Price_Clean'] < lower_bound]):,}")
print(f"  High Outliers:  {len(df[df['Price_Clean'] > upper_bound]):,}")

print("\n\n2. AREA OUTLIERS")
print("-" * 80)

Q1_area = df['Area_Sqft_Clean'].quantile(0.25)
Q3_area = df['Area_Sqft_Clean'].quantile(0.75)
IQR_area = Q3_area - Q1_area
lower_bound_area = Q1_area - 1.5 * IQR_area
upper_bound_area = Q3_area + 1.5 * IQR_area

area_outliers = df[(df['Area_Sqft_Clean'] < lower_bound_area) | (df['Area_Sqft_Clean'] > upper_bound_area)]

print(f"\nArea Bounds (IQR Method):")
print(f"  Lower Bound: {lower_bound_area:>10,.0f} sqft")
print(f"  Upper Bound: {upper_bound_area:>10,.0f} sqft")

print(f"\nArea Outliers Detected:")
print(f"  Total Outliers: {len(area_outliers):,} ({(len(area_outliers)/len(df)*100):.1f}%)")

print("\n\n3. DAYS ON MARKET OUTLIERS")
print("-" * 80)

Q1_dom = df['Days on Market'].quantile(0.25)
Q3_dom = df['Days on Market'].quantile(0.75)
IQR_dom = Q3_dom - Q1_dom
lower_bound_dom = Q1_dom - 1.5 * IQR_dom
upper_bound_dom = Q3_dom + 1.5 * IQR_dom

dom_outliers = df[(df['Days on Market'] < lower_bound_dom) | (df['Days on Market'] > upper_bound_dom)]

print(f"\nDays on Market Bounds (IQR Method):")
print(f"  Lower Bound: {lower_bound_dom:>6.0f} days")
print(f"  Upper Bound: {upper_bound_dom:>6.0f} days")

print(f"\nDays on Market Outliers Detected:")
print(f"  Total Outliers: {len(dom_outliers):,} ({(len(dom_outliers)/len(df)*100):.1f}%)")

print("\n\n4. UNUSUAL OBSERVATIONS - Extreme Properties")
print("-" * 80)

# Highest price properties
print("\nTop 5 Most Expensive Properties:")
top_price = df.nlargest(5, 'Price_Clean')[['Address', 'City', 'Price_Clean', 'Area_Sqft_Clean', 'Bedrooms_Clean', 'Bathrooms_Clean']]
for idx, (_, row) in enumerate(top_price.iterrows(), 1):
    print(f"  {idx}. {row['Address']}")
    print(f"     Price: ${row['Price_Clean']:>12,.2f} | Area: {row['Area_Sqft_Clean']:>7.0f} sqft | {row['Bedrooms_Clean']:.0f}bd/{row['Bathrooms_Clean']:.0f}ba")

# Largest properties
print("\nTop 5 Largest Properties (by Area):")
top_area = df.nlargest(5, 'Area_Sqft_Clean')[['Address', 'City', 'Area_Sqft_Clean', 'Price_Clean', 'Bedrooms_Clean']]
for idx, (_, row) in enumerate(top_area.iterrows(), 1):
    print(f"  {idx}. {row['Address']}")
    print(f"     Area: {row['Area_Sqft_Clean']:>7.0f} sqft | Price: ${row['Price_Clean']:>12,.2f}")

print("\n✓ Market demand and outlier analysis complete")
```

```python
import pandas as pd
import numpy as np

# Load cleaned data
df = pd.read_csv('/content/RealEstate_Dataset.csv')

print("=" * 80)
print("PRICE DISTRIBUTION & MARKET SEGMENTATION ANALYSIS")
print("=" * 80)

print("\n1. PRICE DISTRIBUTION STATISTICS")
print("-" * 80)

percentiles = [10, 20, 30, 40, 50, 60, 70, 80, 90, 95, 99]
print("\nPrice Percentiles:")
for pct in percentiles:
    value = df['Price_Clean'].quantile(pct/100)
    print(f"  {pct:3d}th Percentile: ${value:>12,.2f}")

print("\n\n2. PRICE RANGE DISTRIBUTION")
print("-" * 80)

# Define price ranges
price_ranges = [
    ('$0 - $200K', 0, 200000),
    ('$200K - $400K', 200000, 400000),
    ('$400K - $600K', 400000, 600000),
    ('$600K - $800K', 600000, 800000),
    ('$800K - $1M', 800000, 1000000),
    ('$1M - $1.2M', 1000000, 1200000),
    ('$1.2M - $1.5M', 1200000, 1500000),
]

print("\nPrice Range Distribution:")
for label, low, high in price_ranges:
    count = len(df[(df['Price_Clean'] >= low) & (df['Price_Clean'] <= high)])
    pct = (count / len(df)) * 100
    bar = '█' * int(pct / 2)
    print(f"  {label:20s}: {count:4,} ({pct:5.1f}%) {bar}")

print("\n\n3. MARKET SEGMENTATION")
print("-" * 80)

# Create segments based on price quartiles
df['Price_Segment'] = pd.cut(df['Price_Clean'],
                              bins=[0, 470387, 815630, 1158187.75, 1500000],
                              labels=['Budget', 'Mid-Range', 'Upper-Mid', 'Luxury'],
                              include_lowest=True)

segment_analysis = df.groupby('Price_Segment').agg({
    'Price_Clean': ['count', 'mean', 'median', 'min', 'max'],
    'Area_Sqft_Clean': 'mean',
    'Bedrooms_Clean': 'mean',
    'Bathrooms_Clean': 'mean',
    'Days on Market': 'mean',
    'Property_Type': lambda x: x.mode()[0]
}).round(2)

segment_analysis.columns = ['Count', 'Avg_Price', 'Median_Price', 'Min_Price', 'Max_Price',
                            'Avg_Area', 'Avg_Beds', 'Avg_Baths', 'Avg_DOM', 'Most_Common_Type']

print("\nMarket Segment Characteristics:")
print(segment_analysis.to_string())

print("\n\nSegment Distribution by City:")
segment_by_city = pd.crosstab(df['City'], df['Price_Segment'], margins=True)
print(segment_by_city.to_string())

print("\n\nSegment Distribution by Property Type:")
segment_by_type = pd.crosstab(df['Property_Type'], df['Price_Segment'], margins=True)
print(segment_by_type.to_string())

print("\n\n4. PRICE DISTRIBUTION SKEWNESS")
print("-" * 80)

from scipy.stats import skew, kurtosis

skewness = skew(df['Price_Clean'])
kurt = kurtosis(df['Price_Clean'])

print(f"\nDistribution Characteristics:")
print(f"  Skewness:  {skewness:>7.4f}", end="")
if abs(skewness) < 0.5:
    print(" (Fairly Symmetric)")
elif skewness > 0:
    print(" (Right-skewed - tail extends to higher prices)")
else:
    print(" (Left-skewed - tail extends to lower prices)")

print(f"  Kurtosis:  {kurt:>7.4f}", end="")
if kurt > 0:
    print(" (Leptokurtic - more extreme values than normal)")
else:
    print(" (Platykurtic - fewer extreme values than normal)")

print("\n\n5. SEGMENT DEMAND ANALYSIS")
print("-" * 80)

segment_demand = df.groupby('Price_Segment').agg({
    'Status': lambda x: (x == 'Sold').sum(),
    'Days on Market': 'mean',
}).round(2)

segment_demand.columns = ['Sold_Count', 'Avg_DOM']
segment_demand['Sell_Rate'] = (segment_demand['Sold_Count'] / df.groupby('Price_Segment').size() * 100).round(1)

print("\nSegment Sales Performance:")
print(segment_demand.to_string())

print("\n\n6. PRICE CONCENTRATION ANALYSIS")
print("-" * 80)

# Identify price clustering
price_range = df['Price_Clean'].max() - df['Price_Clean'].min()
std_dev = df['Price_Clean'].std()
cv = (std_dev / df['Price_Clean'].mean()) * 100  # Coefficient of variation

print(f"\nPrice Spread Metrics:")
print(f"  Price Range:          ${price_range:>12,.2f}")
print(f"  Standard Deviation:   ${std_dev:>12,.2f}")
print(f"  Coefficient of Var:   {cv:>13.1f}%")

print(f"\nPrice Distribution Shape:")
print(f"  Narrow range around median suggests uniform market")
print(f"  Wide range with high CV ({cv:.1f}%) suggests diverse market with varied property values")

# Gini coefficient for inequality
sorted_prices = np.sort(df['Price_Clean'].values)
n = len(sorted_prices)
gini = (2 * np.sum(np.arange(1, n+1) * sorted_prices)) / (n * sorted_prices.sum()) - (n + 1) / n
print(f"  Gini Index (Inequality): {gini:>8.4f} (0=perfect equality, 1=perfect inequality)")

print("\n✓ Price distribution and segmentation analysis complete")
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('/content/RealEstate_Dataset.csv')
df['Price_Segment'] = pd.cut(df['Price_Clean'],
                              bins=[0, 470387, 815630, 1158187.75, 1500000],
                              labels=['Budget', 'Mid-Range', 'Upper-Mid', 'Luxury'],
                              include_lowest=True)

# Create detailed comparison visualizations
fig, axes = plt.subplots(2, 2, figsize=(16, 12), dpi=200)
fig.patch.set_facecolor('#f8f9fa')

# 1. City Market Overview - Dual Axis
ax1 = axes[0, 0]
cities = df['City'].unique()
city_data = df.groupby('City').agg({
    'Price_Clean': 'mean',
    'Days on Market': 'mean'
}).sort_values('Price_Clean', ascending=False)

ax1_2 = ax1.twinx()
bars = ax1.bar(range(len(city_data)), city_data['Price_Clean']/1e6, alpha=0.7, color='#3498db', label='Avg Price')
line = ax1_2.plot(range(len(city_data)), city_data['Days on Market'], 'o-', color='#e74c3c', linewidth=2.5,
                   markersize=10, label='Avg DOM')
ax1.set_xticks(range(len(city_data)))
ax1.set_xticklabels(city_data.index, fontsize=10, fontweight='bold')
ax1.set_ylabel('Average Price (Millions $)', fontsize=11, fontweight='bold', color='#3498db')
ax1_2.set_ylabel('Average Days on Market', fontsize=11, fontweight='bold', color='#e74c3c')
ax1.set_title('City Market Overview: Price vs Sales Speed', fontsize=13, fontweight='bold', pad=15)
ax1.tick_params(axis='y', labelcolor='#3498db')
ax1_2.tick_params(axis='y', labelcolor='#e74c3c')
ax1.grid(axis='y', alpha=0.3)
ax1.legend(loc='upper left', fontsize=10)
ax1_2.legend(loc='upper right', fontsize=10)

# 2. Property Type - Price vs Availability
ax2 = axes[0, 1]
ptype_stats = df.groupby('Property_Type').agg({
    'Price_Clean': 'mean',
    'Property_Type': 'count'
}).rename(columns={'Property_Type': 'Count'}).sort_values('Price_Clean', ascending=False)

scatter_colors = ['#3498db', '#e74c3c', '#2ecc71', '#f39c12', '#9b59b6']
scatter = ax2.scatter(ptype_stats['Count'], ptype_stats['Price_Clean']/1e6,
                     s=300, c=scatter_colors, alpha=0.7, edgecolors='black', linewidth=2)
for idx, (ptype, row) in enumerate(ptype_stats.iterrows()):
    ax2.annotate(ptype, (row['Count'], row['Price_Clean']/1e6),
                fontsize=10, fontweight='bold', ha='center', va='center')
ax2.set_xlabel('Number of Listings', fontsize=11, fontweight='bold')
ax2.set_ylabel('Average Price (Millions $)', fontsize=11, fontweight='bold')
ax2.set_title('Property Type: Market Presence vs Price Point', fontsize=13, fontweight='bold', pad=15)
ax2.grid(True, alpha=0.3)

# 3. Bedroom Analysis
ax3 = axes[1, 0]
bed_analysis = df.groupby('Bedrooms_Clean')['Price_Clean'].agg(['mean', 'count', 'std'])
x_pos = np.arange(len(bed_analysis))
bars = ax3.bar(x_pos, bed_analysis['mean']/1e6, yerr=bed_analysis['std']/1e6,
               alpha=0.7, color='#2ecc71', capsize=5, error_kw={'linewidth': 2}, edgecolor='black', linewidth=1.5)
ax3.set_xticks(x_pos)
ax3.set_xticklabels([f'{int(x)}' for x in bed_analysis.index], fontsize=10, fontweight='bold')
ax3.set_xlabel('Number of Bedrooms', fontsize=11, fontweight='bold')
ax3.set_ylabel('Average Price (Millions $)', fontsize=11, fontweight='bold')
ax3.set_title('Price Analysis by Bedroom Count (with Std Dev)', fontsize=13, fontweight='bold', pad=15)
ax3.grid(axis='y', alpha=0.3)
for i, (bar, count) in enumerate(zip(bars, bed_analysis['count'])):
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width()/2., height + 0.05, f'n={int(count)}',
            ha='center', va='bottom', fontsize=9, fontweight='bold')

# 4. Status Trend Analysis
ax4 = axes[1, 1]
status_metrics = df.groupby('Status').agg({
    'Status': 'count',
    'Price_Clean': 'mean',
    'Days on Market': 'mean'
}).rename(columns={'Status': 'Count'})

status_order = ['For Sale', 'Pending', 'Sold']
status_metrics = status_metrics.reindex(status_order)
status_colors = {'For Sale': '#e74c3c', 'Pending': '#f39c12', 'Sold': '#2ecc71'}

# Create grouped bar chart
x = np.arange(len(status_metrics))
width = 0.25

bars1 = ax4.bar(x - width, status_metrics['Count']/100, width, label='Listings (÷100)',
               color='#95a5a6', alpha=0.8, edgecolor='black', linewidth=1.5)
ax4_2 = ax4.twinx()
bars2 = ax4_2.bar(x, status_metrics['Price_Clean']/1e6, width, label='Avg Price',
                 color=[status_colors[status] for status in status_order], alpha=0.8, edgecolor='black', linewidth=1.5)
bars3 = ax4_2.bar(x + width, status_metrics['Days on Market'], width, label='Days on Market',
                 color='#34495e', alpha=0.8, edgecolor='black', linewidth=1.5)

ax4.set_xlabel('Property Status', fontsize=11, fontweight='bold')
ax4.set_ylabel('Listing Count', fontsize=11, fontweight='bold')
ax4_2.set_ylabel('Price (Millions $) & DOM (Days)', fontsize=11, fontweight='bold')
ax4.set_xticks(x)
ax4.set_xticklabels(status_order, fontsize=10, fontweight='bold')
ax4.set_title('Property Status Analysis: Volume, Price & DOM', fontsize=13, fontweight='bold', pad=15)
ax4.grid(axis='y', alpha=0.3)

# Combined legend
handles1 = [bars1]
handles2 = [bars2, bars3]
ax4.legend(handles1, ['Listings'], loc='upper left', fontsize=10)
ax4_2.legend(handles2, ['Avg Price', 'DOM'], loc='upper right', fontsize=10)

plt.tight_layout()
plt.savefig('/content/02_Detailed_Market_Analysis.png', dpi=200, bbox_inches='tight')
print("✓ Saved: 02_Detailed_Market_Analysis.png")
plt.close()

# Create Price Segmentation Deep Dive
fig, axes = plt.subplots(2, 2, figsize=(16, 12), dpi=200)
fig.patch.set_facecolor('#f8f9fa')

segment_colors = {'Budget': '#3498db', 'Mid-Range': '#2ecc71', 'Upper-Mid': '#f39c12', 'Luxury': '#e74c3c'}
segments = ['Budget', 'Mid-Range', 'Upper-Mid', 'Luxury']

# 1. Segment Size and Sales
ax1 = axes[0, 0]
segment_stats = df.groupby('Price_Segment').agg({
    'Price_Clean': ['count', lambda x: (x.index.isin(df[df['Status']=='Sold'].index)).sum()]
}).rename(columns={'<lambda_0>': 'Sold'})
segment_stats.columns = ['Total', 'Sold']
segment_stats = segment_stats.reindex(segments)

x = np.arange(len(segment_stats))
width = 0.35
bars1 = ax1.bar(x - width/2, segment_stats['Total'], width, label='Total Listings',
               color='#95a5a6', alpha=0.8, edgecolor='black', linewidth=1.5)
bars2 = ax1.bar(x + width/2, segment_stats['Sold'], width, label='Sold',
               color=[segment_colors[seg] for seg in segments], alpha=0.8, edgecolor='black', linewidth=1.5)
ax1.set_ylabel('Count', fontsize=11, fontweight='bold')
ax1.set_xticks(x)
ax1.set_xticklabels(segments, fontsize=10, fontweight='bold')
ax1.set_title('Market Segment Size & Sales Volume', fontsize=13, fontweight='bold', pad=15)
ax1.legend(fontsize=10)
ax1.grid(axis='y', alpha=0.3)
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2., height, f'{int(height)}',
                ha='center', va='bottom', fontsize=9, fontweight='bold')

# 2. Segment Price Distribution
ax2 = axes[0, 1]
for segment in segments:
    seg_data = df[df['Price_Segment'] == segment]['Price_Clean']/1e6
    ax2.hist(seg_data, bins=20, alpha=0.6, label=segment, color=segment_colors[segment], edgecolor='black')
ax2.set_xlabel('Price (Millions $)', fontsize=11, fontweight='bold')
ax2.set_ylabel('Frequency', fontsize=11, fontweight='bold')
ax2.set_title('Price Distribution by Segment', fontsize=13, fontweight='bold', pad=15)
ax2.legend(fontsize=10)
ax2.grid(axis='y', alpha=0.3)

# 3. Segment Characteristics
ax3 = axes[1, 0]
seg_chars = df.groupby('Price_Segment').agg({
    'Area_Sqft_Clean': 'mean',
    'Bedrooms_Clean': 'mean',
    'Bathrooms_Clean': 'mean'
}).reindex(segments)

x = np.arange(len(seg_chars))
width = 0.25
ax3.bar(x - width, seg_chars['Area_Sqft_Clean']/100, width, label='Area (÷100 sqft)', color='#3498db', alpha=0.8, edgecolor='black', linewidth=1.5)
ax3.bar(x, seg_chars['Bedrooms_Clean'], width, label='Bedrooms', color='#2ecc71', alpha=0.8, edgecolor='black', linewidth=1.5)
ax3.bar(x + width, seg_chars['Bathrooms_Clean'], width, label='Bathrooms', color='#e74c3c', alpha=0.8, edgecolor='black', linewidth=1.5)
ax3.set_ylabel('Count', fontsize=11, fontweight='bold')
ax3.set_xticks(x)
ax3.set_xticklabels(segments, fontsize=10, fontweight='bold')
ax3.set_title('Segment Characteristics: Area, Beds & Baths', fontsize=13, fontweight='bold', pad=15)
ax3.legend(fontsize=10)
ax3.grid(axis='y', alpha=0.3)

# 4. Segment Performance Metrics
ax4 = axes[1, 1]
seg_perf = df.groupby('Price_Segment').agg({
    'Days on Market': 'mean',
}).reindex(segments)
seg_perf['Sell_Rate'] = df.groupby('Price_Segment').apply(lambda x: (x['Status'] == 'Sold').sum() / len(x) * 100)
seg_perf = seg_perf.reindex(segments)

x = np.arange(len(seg_perf))
width = 0.35
ax4_2 = ax4.twinx()

bars1 = ax4.bar(x - width/2, seg_perf['Days on Market'], width, label='Avg DOM',
               color='#95a5a6', alpha=0.8, edgecolor='black', linewidth=1.5)
bars2 = ax4_2.bar(x + width/2, seg_perf['Sell_Rate'], width, label='Sell-Through Rate',
                 color=[segment_colors[seg] for seg in segments], alpha=0.8, edgecolor='black', linewidth=1.5)

ax4.set_ylabel('Average Days on Market', fontsize=11, fontweight='bold')
ax4_2.set_ylabel('Sell-Through Rate (%)', fontsize=11, fontweight='bold')
ax4.set_xticks(x)
ax4.set_xticklabels(segments, fontsize=10, fontweight='bold')
ax4.set_title('Segment Sales Performance', fontsize=13, fontweight='bold', pad=15)
ax4.grid(axis='y', alpha=0.3)

# Combined legend
handles = [bars1, bars2]
labels = ['Avg DOM', 'Sell-Through Rate']
ax4.legend(handles, labels, loc='upper left', fontsize=10)

plt.tight_layout()
plt.savefig('/content/03_Price_Segmentation_Analysis.png', dpi=200, bbox_inches='tight')
print("✓ Saved: 03_Price_Segmentation_Analysis.png")
plt.close()

print("\n✓ Detailed analysis visualizations complete")
```
